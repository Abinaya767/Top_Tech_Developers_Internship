```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,30,40,50]
z = [15,25,35,45,55]

plt.plot(x,y,label="X vs Y")
plt.plot(x,z,label="X vs Z")

plt.xlabel("X Values")
plt.ylabel("Y and Z Values")
plt.title("Comparison Graph")

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,30,40,50]

size = [100,200,300,400,500]

plt.scatter(x,y,s=size)

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Bubble Chart")

plt.show()
```

```python
import matplotlib.pyplot as plt

stages = ["Visit","Signup","Login","Purchase"]

y = [4,3,2,1]

plt.fill_betweenx([4.3,3.7], -50, 50, label="Visit")
plt.fill_betweenx([3.3,2.7], -40, 40, label="Signup")
plt.fill_betweenx([2.3,1.7], -30, 30, label="Login")
plt.fill_betweenx([1.3,0.7], -20, 20, label="Purchase")

plt.yticks(y, stages)

plt.xlabel("Users")
plt.ylabel("Stages")

plt.title("Funnel Chart")

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

levels = ["Top","Middle","Lower","Bottom"]
y = [4,3,2,1]


plt.fill_betweenx([4.5,3.5], -10, 10)
plt.fill_betweenx([3.5,2.5], -20, 20)
plt.fill_betweenx([2.5,1.5], -30, 30)
plt.fill_betweenx([1.5,0.5], -40, 40)

plt.yticks(y, levels)

plt.xlabel("Values")
plt.ylabel("Levels")

plt.title("Pyramid Chart")

plt.show()
```
