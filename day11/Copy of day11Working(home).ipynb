```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Line Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["A","B","C","D"]
y = [20,40,60,80]

plt.bar(x,y)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50]
y = [15,25,35,45,55]

plt.scatter(x,y)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

values = [20,30,25,25]
labels = ["A","B","C","D"]

plt.pie(values, labels=labels)
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,30,40,50]

print("X Values :", x)
print("Y Values :", y)

plt.plot(x,y)

plt.xlim(0,6)
plt.ylim(0,60)

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Line Graph")

plt.show()
```
