```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/student-scores_DATASET(cleaned).csv')  # change file name if needed
df.head()
```

```python
import matplotlib.pyplot as plt

subjects = df.columns[1:8]   # adjust if needed based on your dataset

avg_marks = df[subjects].mean()

plt.figure(figsize=(10,5))
plt.bar(avg_marks.index, avg_marks.values)

plt.title("Average Marks by Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")

plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

gender_avg = df.groupby('Gender')['Total_Marks'].mean()

plt.figure(figsize=(6,4))
plt.plot(gender_avg.index, gender_avg.values, marker='o', color='purple')

plt.title("Gender-wise Average Marks")
plt.xlabel("Gender")
plt.ylabel("Average Total Marks")

plt.show()
```

```python
grade_counts = df['Grade'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%')

plt.title("Grade Distribution")
plt.show()
```
