```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,30,40,50]

print("X Values :", x)
print("Y Values :", y)

plt.plot(x,y)

plt.xlim(0,10)
plt.ylim(0,60)

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Line Graph")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,30,40,50]

a = [1,2,3,4,5]
b = [15,25,35,45,55]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlim(0,6)
plt.ylim(0,60)
plt.title("Line Graph")

plt.subplot(1,2,2)
plt.bar(a,b)
plt.xlim(0,6)
plt.ylim(0,60)
plt.title("Bar Graph")

plt.show()
```

```python
import matplotlib.pyplot as plt
import random
import numpy as np

x = random.randint(1,50)
y = random.randint(1,50)

a = x * np.sin(x)
b = y * np.cos(y)

plt.plot(a,b,"o")

plt.xlabel("Sin Values")
plt.ylabel("Cos Values")
plt.title("Sin and Cos Graph")

plt.show()
```

```python
import matplotlib.pyplot as plt

values = [20,30,25,25]
labels = ["A","B","C","D"]

plt.pie(values, labels=labels)

plt.title("Pie Chart")

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [10,20,20,30,30,30,40,40,50]

plt.hist(data, rwidth=10)

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7]
y = [10,25,15,40,30,50,45]

plt.stem(x,y)

plt.title("Stem Plot")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.step(x,y)

plt.title("Step Graph")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.fill_between(x,y)

plt.title("Area Graph")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]

y1 = [10,20,30,40,50]
y2 = [5,10,15,20,25]

plt.stackplot(x,y1,y2)

plt.title("Stack Plot")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8]
y = [10,20,15,25,30,35,40,45]

plt.hexbin(x,y)

plt.title("Hexbin Graph")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y = x

plt.polar(x,y)

plt.title("Polar Graph")

plt.show()
```

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]
y1 = [10,20,30,40,50]
z1 = [15,25,35,45,55]

plt.subplot(1,2,1)
plt.plot(x1,y1,label="X1 vs Y1")

plt.title("Line Chart")
plt.legend()

plt.subplot(1,2,2)

plt.bar(x1,z1,label="X1 vs Z1")
plt.title("Bar Chart")
plt.legend()

plt.show()
```
